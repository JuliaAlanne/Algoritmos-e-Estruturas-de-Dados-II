**2.2- 60% of the emails hitting my inbox is spam. You design a phenomenal spam filter which is able to tell me, with 98% accuracy,
whether an email is spam or not: if an email is not spam, the system has a 98% probability of saying so. The filter knows 60% of
emails are spam and so it will flag 60% of my emails. Suppose
that, at the end of the week, I look in my spam box and see 963
emails. Use Bayes’ Theorem to calculate how many of those 963
emails in my spam box I should suspect to be non-spam.**


**2.2** Resposta:

Queremos saber a probabilidade do filtro sinalizar um e-mail como Spam e não ser Spam, P(Não Spam|Sinalizado). Sabemos que:
P(Spam) = 0,6; P(Sinalizado)= 0,6
; P(Não Spam)= 1- P(Spam)= 1-0,6 = 0,4; P(Não Sinalizado|Não Spam)= 0,98

Então, P (Sinalizado|Não Spam)=1- P(Não Sinalizado|Não Spam)= 1-0,98= 0,02

Pelo teorema de Bayes, temos que:

P(Não Spam|Sinalizado) = P(Sinalizado|Não Spam) x P(Não Spam) /P (Sinalizado) = 0,02 x 0,4 / 0,6 = 0,013333

  Tendo em vista que tinha 963 e-mails na caixa de Spam, então 963 * P(Não Spam|Sinalizado) = 12,84

Aproximadamente 13 dos e-mails na caixa de Spam não são Spam.

**2.5. How many bits do we need to independently encode v1 and v2
from http://www.networkatlas.eu/exercises/2/5/data.txt?
How much would we save in encoding v1 if we knew v2?**

**2.5**Resposta:

Para codificar uma variavel, o número de bits necessários depende da entropia da variável.

A entropia de X é : $H(X) = - \sum p(x) \log_2 p(x)$.

Para codificar duas variáveis v1 e v2 de forma independente, calculamos a entropia de cada uma e somamos, já que assumimos que as duas variáveis não influenciam uma à outra. Logo:
$H(v1,v2)=H(v1)+H(v2)$

Com a soma das entropias podemos saber o número total de bits necessários para codificar v1 e v2 independentemente.

A diferença entre a entropia condicional (H(v2, v2)) e a entropia de (v2) nos diz quantos bits ainda precisamos para codificar v1 ao conhecer v2.

$H(v1|v2)=H(v1,v2)-H(v2)$

A economia de bits que ocorre quando conhecemos v2 ao codificar v1 é a diferença entre a entropia de  v1 isoladamente e a entropia condicional:

$Economia= H(v1)-H(v1,v2)$

Com os dados contamos a ocorrencia de cada elemento nos vetores para calcular a probabilidade.como os vetores v1 e v2 possuem 30 elementos cada, temos que:

Probabilidades de v1:
$P(a)=8/30, P(b)=6/30, P(c)=8/30,P(d)=8/30,P(e)=11/30,P(f)=5/30$

Probabilidades de v2:
$P(a)=5/30, P(d)=4/30, P(f)=6/30,P(g)=5/30,P(h)=5/30,P(j)=6/30, P(k)=1/30,P(s)=8/30$

**4.3. The network in http://www.networkatlas.eu/exercises/4/3/
data.txt is a hypergraph, with a hyperedge per line. Transform it
in a unipartite network in which each hyperedge is split in edges
connecting all nodes in the hyperedge. Then transform it into a
bipartite network in which each hyperedge is a node of one type
and its nodes connect to it.**

**4.3** Resposta:

Rede Unipartida: Conecta todos os nós de cada hiperaresta entre si, criando uma rede onde todos os nós estão interligados com base nas hiperarestas.

Rede Bipartida: Cria um novo nó para cada hiperaresta e conecta os nós que pertencem a essa hiperaresta a esse novo nó, resultando em uma separação clara entre os tipos de nós.

Precisamos examinar cada linha do arquivo que representa as hiperarestas. Cada linha conecta um grupo de nós.

Para cada hiperaresta, conecte todos os nós entre si. Isso significa que se uma hiperaresta conecta os nós A, B e C, você deve criar as arestas: A - B, A - C, B - C. E o resultado é uma rede unipartida onde as arestas representam as conexões diretas entre todos os nós que pertencem a cada hiperaresta.

Para transformar a rede unipartida em Rede Bipartida

* Identificamos as Hiperarestas com base as hiperarestas definidas no arquivo.
* Para cada hiperaresta, criamos um novo nó que a representa. Por exemplo, se a primeira hiperaresta conecta os nós A, B e C, criamos um novo nó chamado H1 para representar essa hiperaresta.
* Conectamos todos os nós da hiperaresta (A, B, C) ao novo nó (H1). Isso significa que o nó H1 terá arestas conectadas a A, B e C.

Temos como resultado uma rede bipartida onde um conjunto contém os nós da hiperaresta (A, B, C), e o outro conjunto contém os nós que representam as hiperarestas (H1, H2, etc.).
